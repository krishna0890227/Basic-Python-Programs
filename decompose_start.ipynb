```python
# first order exponential smoothing
def single_exp_smoothing(x, alpha):
    print(type(x))
    F= [x[0]] # first value os same as series
    for t in range(1, len(x)):
        F.append(alpha * x[t] + (1-alpha)*F[t-1])
    return F
```

```python
# second order exponential smoothing
def double_exp_smoothing(x, alpha, beta):
    yhat=[x[0]] # first value is same as series
    for t in range(1, len(x)):
        if t==1:
            F, T=x[0], x[1]-x[0]
        F_n_1, F= F, alpha*x[t] + (1-alpha)*(F+T)
        T= beta*(F-F_n_1)+(1-beta)*T
        yhat.append(F+T)
    return yhat
```

```python
workdata['first_exp']=single_exp_smoothing(workdata['Electric_load'],0.7)
fig=plt.figure(figsize=(5.5, 5.5))
ax=fig.add_subplot(2,1,1)
workdata['Electric_load'].plot(ax=ax, color='b')
ax.set_title("Electric Load")
ax=fig.add_subplot(2,1,2)
workdata['first_exp'].plot(ax=ax, color='r')
ax.set_title('Result of single exponential smoothing')
plt.tight_layout()
```

```python

from pandas import Series
from matplotlib import pyplot
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
 

def predict(coef, history):
    yhat=0.0
    for i in range(1,len(coef)+1):
        yhat +=coef[i-1]*history[-i]
        
    return yhat
series = arma_work.iloc[:,0] #Series.from_csv('daily-minimum-temperatures.csv', header=0)
X = series.values
#print(arma_work.head())
size = len(X) - 24
train, test = X[0:size], X[size:]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
	model = ARIMA(history, order=(1,0,1))
	model_fit = model.fit(trend='nc', disp=False)
	ar_coef, ma_coef = model_fit.arparams, model_fit.maparams
	resid = model_fit.resid
	yhat = predict(ar_coef, history) + predict(ma_coef, resid)
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)
	#print('>predicted=%.3f, expected=%.3f' % (yhat, obs))
rmse = sqrt(mean_squared_error(test, predictions))
plt.plot(test,'b--')
plt.plot(predictions,'r-.')
plt.show()
arima_mape=mape(test, predictions)
print("Test RMSE: %.3f and Test MAPE: %.3f" %(rmse, arima_mape))
```

```python
def mape(actual,pred):
    if (not len(actual) == len(pred) or len(actual) == 0):
        return -1.0
    total = 0.0
    cons = 0.1
    for x in range(len(actual)):
         if (actual[x] <=cons):
            total += abs((actual[x]-pred[x])/(cons+actual[x]))
         else: 
            total += abs((actual[x]-pred[x])/actual[x])*100
    return total/len(actual)
```

```python
import math
def Rmse(actual,pred):
	if (not len(actual) == len(pred) or len(actual) == 0):
		return -1.0
	sumSquares = 0.0
	for x in range(len(actual)):
		sumSquares += math.pow(pred[x]-actual[x],2.0)
	return math.sqrt(sumSquares/len(actual))
```
